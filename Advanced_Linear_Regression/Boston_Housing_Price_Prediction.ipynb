```python
from sklearn.model_selection import train_test_split

target=data.PRICE
features = data.drop("PRICE",axis=1)

X_train,X_test,y_train,y_test=train_test_split(features,target,train_size=0.8,random_state=10)
```

```python
X_train.head()
```

```python
X_train.shape,X_test.shape
```

```python
from sklearn.linear_model import LinearRegression

reg=LinearRegression()
reg.fit(X_train,y_train)
```

```python
print("Trainig R squared", reg.score(X_train, y_train))
print("Test R squared", reg.score(X_test, y_test))
```

```python
import statsmodels.api as sm

y= y_train
mod=sm.OLS(y,X_train)
res=mod.fit()
```

```python
res.summary()
```

```python
from statsmodels.stats.diagnostic import het_goldfeldquandt

het_goldfeldquandt(y_train, X_train)
```

```python
from statsmodels.stats.outliers_influence import variance_inflaction_factor

mc=[]

for i in range(X_train.shape[1]):
    mc.append(variance_inflation_factor(X_train.values,i))
    
mc=pd.Series(mc,index=X_train.columns)
mc=round(mc,2)
```
